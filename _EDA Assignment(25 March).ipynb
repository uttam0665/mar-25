```python
"""Q1. Load the flight price dataset and examine its dimensions. How many rows and columns does the
dataset have?

Ans:-
"""
import pandas as pd

# Load the dataset
df=pd.read_excel("flight_price.xlsx")
num_rows, num_columns=df.shape
print("Number of rows:", num_rows)
print("Number of columns:", num_columns)
```

```python
"""Q2. What is the distribution of flight prices in the dataset? Create a histogram to visualize the
distribution."""

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df=pd.read_excel("flight_price.xlsx")


flight_prices = df['Price']  

# Create a histogram
plt.figure(figsize=(10, 6))
plt.hist(flight_prices, bins=20, color='blue', alpha=0.7)

# Add labels and title
plt.xlabel('Flight Price')
plt.ylabel('Frequency')
plt.title('Distribution of Flight Prices')

# Show the plot
plt.show()
```

```python
"""Q3. What is the range of prices in the dataset? What is the minimum and maximum price?"
Ans:=
"""
import pandas as pd

flight_price_data = pd.read_excel("flight_price.xlsx")  
flight_prices = flight_price_data['Price'] 
price_range = flight_prices.max() - flight_prices.min()
minimum_price = flight_prices.min()
maximum_price = flight_prices.max()

print("Price Range:", price_range)
print("Minimum Price:", minimum_price)
print("Maximum Price:", maximum_price)
```

```python
"""Q4. How does the price of flights vary by airline? Create a boxplot to compare the prices of different
airlines.
Ans:-
"""
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

flight_price_data = pd.read_excel('flight_price.xlsx')

plt.figure(figsize=(12, 6))
sns.boxplot(x='Duration', y='Price', data=flight_price_data)


plt.xlabel('Duration')
plt.ylabel('Price')
plt.title('Flight Prices by Airline')

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
"""Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.
Ans:=
"""
import pandas as pd

df = pd.read_csv('googleplaystore.csv')

# Convert the price column to a float
df['Price'] = df['Price'].astype('float')

# Group the data by app type and calculate the average price
grouped_df = df.groupby('Rating')['Price'].mean()

# Create a bar chart
plt.bar(grouped_df.index, grouped_df.values)
plt.title('Average App Price by Type')
plt.xlabel('App Type')
plt.ylabel('Average Price (USD)')
plt.show()

```
